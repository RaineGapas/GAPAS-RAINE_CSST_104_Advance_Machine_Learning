```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Lesson_4_student_grades_dataset.csv')
print(df)
```

```python
plt.scatter(df['Hours_of_Study'],df['Grades'], color='pink')
plt.xlabel('Hours_of_Study(x)')
plt.ylabel('Grades(y)')
plt.title('Scatter PLot of Data')
```

```python
x = df[['Hours_of_Study']] #Independent variable
y = df[['Grades']] #dependent variable
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print('slope', slope)
print('intercept', intercept)
```

```python
plt.scatter(x,y , color = 'pink')

plt.plot(x,model.predict(x), color = 'magenta')
plt.xlabel('INDEPENDENT VARIABLE (x)')
plt.ylabel('DEPENDENT VARIABLE (y)')
plt.title('LINEAR REGRESSION')
plt.show()
```

```python
new_x =[[np.random.randint(10,60)]]
predicted_y = model.predict(new_x)
print('Prediction of Grade:', predicted_y)
```
